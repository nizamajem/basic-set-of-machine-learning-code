```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
from pandas.io.parsers import read_csv
from google.colab import files
CSV = files.upload()
```

```python
import io
df = pd.read_csv(io.StringIO(CSV['Iris.csv'].decode('utf-8')))
```

```python
df.head()
```

```python
df = df.drop(columns='Id')
```

```python
category = pd.get_dummies(df.Species)
category
```

```python
new_df = pd.concat([df, category], axis=1)
new_df
```

```python
new_df = new_df.drop(columns='Species')
new_df
```

```python
# mengkonversi datafram menjadi numpy array dengan fungsi values 

dataset = new_df.values
dataset
```

```python
# Pilih 4 kolom pertama untuk dijadikan sebagai atribut
X = dataset[:,0:4]
# Pilih 3 kolom terakhir sebagai label
y = dataset[:,4:7]
```

```python
# normlaization 
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
X_scale
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_scale, y, test_size=0.3)
```

```python
# arsitektur menggunakan 3 layer
model = Sequential([    
                    Dense(64, activation='relu', input_shape=(4,)),    
                    Dense(64, activation='relu'),    
                    Dense(3, activation='softmax'),])
```

```python
#optiizer lossnya 
model.compile(optimizer='Adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
hist = model.fit(X_train, Y_train, epochs=100)
```

```python
model.evaluate(X_test, Y_test)
```

```python
# Plot Loss dan Akurasi dari Trained Model
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(hist.history['loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train'], loc='upper right')
plt.show()
```

```python
plt.plot(hist.history['accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train'], loc='upper left')
plt.show()
```

```python
# Penggunaan Callbacks
import tensorflow as tf
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('accuracy')>0.9):
      print("\nAkurasi telah mencapai >90%!")
      self.model.stop_training = True
callbacks = myCallback()
```

```python
model.fit(X_train, Y_train, epochs=50, callbacks=[callbacks])
```
